This data was pulled from a research dataset on Dyslexia test results. This research project focused on identifying Dyslexia in individuals using eye scanning techniques and tracking four different eye movements over time: sacades, fixations, sweeping movements, and transients (see article below).

https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0165508

The study consisted of 143 male test subjects, 68 of whom were not dyslexic and 75 of whom were dyslexic, and 38 female test subjects, 18 of whom were not dyslexic and 20 of whom were dyslexic.

The goal of this project is to convert each of the total 181 datasets into an array consisting of data from the four different eye movements over time that contribute to the demonstrated reading patterns, and fit a neural network to that input data and predict whether the data correpsonds to that of a dyslexic or non-dyslexic person.

```
#storing group data
Category4_data=pd.DataFrame(data=Category4)
Category3_data=pd.DataFrame(data=Category3)
Category2_data=pd.DataFrame(data=Category2)
Category_data=pd.DataFrame(data=Category)
categories=pd.concat([Category4_data,Category3_data,Category2_data,Category_data],axis=0) #combining dataframes into one
from sklearn import preprocessing
#converting character string into integer value (0,1)
category=preprocessing.LabelEncoder()
y=categories
y2=y.apply(category.fit_transform)
```

```
#creating input arrays from lists
LX_tot=np.array(arrays_LX)
LY_tot=np.array(arrays_LY)
RX_tot=np.array(arrays_RX)
RY_tot=np.array(arrays_RY)
array_tot=np.concatenate((LX_tot,LY_tot,RX_tot,RY_tot), axis=1)
#now have 181 rows with 4000 columns
```

```
from sklearn.model_selection import train_test_split
#creating training and testing sub-groups for network
array_tot_train,array_tot_test, y2_train, y2_test=train_test_split(array_tot, y2, test_size=0.4, random_state=3)
#LX_tot_train, LX_tot_test, LY_tot_train, LY_tot_test, RX_tot_train, RX_tot_test, RY_tot_train, RY_tot_test, y2_train, y2_test = train_test_split(LX_tot, LY_tot, RX_tot, RY_tot, y2, test_size=0.3, random_state=3)
```

```
from sklearn.preprocessing import StandardScaler
#scaling data to get more accurate predictions
data_scaler=StandardScaler()
data_scaler.fit(array_tot_train)
array_tot_train=data_scaler.transform(array_tot_train)
array_tot_test=data_scaler.transform(array_tot_test)
"""
data_scaler.fit(LX_tot_train)
LX_tot_train=data_scaler.transform(LX_tot_train)
LX_tot_test=data_scaler.transform(LX_tot_test)
data_scaler.fit(LY_tot_train)
LY_tot_train=data_scaler.transform(LY_tot_train)
LY_tot_test=data_scaler.transform(LY_tot_test)
data_scaler.fit(RX_tot_train)
RX_tot_train=data_scaler.transform(RX_tot_train)
RY_tot_test=data_scaler.transform(RY_tot_test)
"""
```

```
from sklearn.neural_network import MLPClassifier
#setting up network
obj_neural=MLPClassifier(hidden_layer_sizes=(50,50,50), max_iter=25000, solver='lbfgs')
obj_neural.fit(array_tot_train, y2_train.values.ravel())
predictions=obj_neural.predict(array_tot_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y2_test, predictions))
#compares predicted classes to actual classes
#--> showing which objects got classified the right way according to a matrix.
print(classification_report(y2_test, predictions)) #analyzes the predictions
```
